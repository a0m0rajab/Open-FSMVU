```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python

from sklearn.datasets import load_boston
boston = load_boston()
```

```python
print(boston.DESCR)
```

```python
data = pd.DataFrame(boston.data, columns=boston["feature_names"])
print(data.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test =  train_test_split(data,  boston.target , test_size = 0.33, random_state = 5)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lrm = LinearRegression()
lrm.fit(X_train, Y_train)

Y_pred = lrm.predict(X_test)
plt.figure(figsize=(20,10))
plt.scatter(Y_test, Y_pred)
plt.xlabel("Prices")
plt.ylabel("Predicted prices")
plt.title("Prices vs Predicted prices")
```

```python
mse = sklearn.metrics.mean_squared_error(Y_test, Y_pred)
print(mse)
```

```python
def GDMSE(test,pred,learningRate,iteration=100,theta=100):
    history1 = [x for x in range(iteration)]
    history1[0]=theta;
    for i in range(100):
        theta = theta - learningRate * sklearn.metrics.mean_squared_error(test, pred)
        history1[i]=theta;
        print(i,theta)
    plt.figure(figsize=(20,10))
    plt.scatter(history1, [x for x in range(iteration)])
```

```python
GDMSE(Y_test,Y_pred,0.5)

```

```python
GDMSE(Y_test,Y_pred,0.1)
```

```python
GDMSE(Y_test,Y_pred,0.02)
```
